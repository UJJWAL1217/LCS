# Loading the dataset into notebook

```python
import pandas as pd

# Load the dataset into a DataFrame
file_path = 'books.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()
```

# Q1: How many rows and columns are there in books.csv dataset?

```python
# Get the number of rows and columns in the DataFrame
num_rows, num_cols = df.shape

# Print the number of rows and columns
print(f"The books.csv dataset has {num_rows} rows and {num_cols} columns.")
```

# Q3: How many unique books are present in the dataset ? Evaluate based on the 'book_id' after removing null values in the original_title column.

```python
# Remove rows with null values in the original_title column
df_cleaned = df.dropna(subset=['original_title'])

# Count the number of unique books based on book_id
num_unique_books = df_cleaned['book_id'].nunique()

# Print the number of unique books
print(f"There are {num_unique_books} unique books in the dataset after removing null values in the original_title column.")
```

```python
average_rating_all_books = df['average_rating'].mean()

# Create a bar plot with the average rating as an annotation
plt.figure(figsize=(6, 4))
plt.bar(['Average Rating'], [average_rating_all_books], color='lightgreen')
plt.text(0, average_rating_all_books + 0.05, f'{average_rating_all_books:.2f}', ha='center')
plt.title('Average Rating of All Books in the Dataset')
plt.ylabel('Average Rating')
plt.ylim(0, 5)  # Set y-axis limit to the range of ratings (0 to 5)
plt.show()
```

# Q6: Which book (title) has the maximum number of ratings based on â€˜work_ratings_countâ€™.

```python
# Find the index of the row with the maximum value in the 'work_ratings_count' column
max_ratings_index = df['work_ratings_count'].idxmax()

# Get the title of the book with the maximum number of ratings
book_with_max_ratings = df.loc[max_ratings_index, 'title']

# Print the title of the book with the maximum number of ratings
print(f"The book with the maximum number of ratings is '{book_with_max_ratings}' with {df.loc[max_ratings_index, 'work_ratings_count']} ratings.")
```

# Q6: Which book (title) has the maximum number of ratings based on â€˜work_ratings_countâ€™.(Visual Representation)

```python
# Sort the DataFrame by 'work_ratings_count' in descending order
df_sorted_by_ratings = df.sort_values(by='work_ratings_count', ascending=False)

# Get the top 10 books with the highest ratings count
top_books = df_sorted_by_ratings.head(10)

# Create a bar plot to visualize the ratings count for the top 10 books
plt.figure(figsize=(10, 6))
plt.bar(top_books['title'], top_books['work_ratings_count'], color='skyblue')
plt.title('Top 10 Books with Highest Ratings Count')
plt.xlabel('Book Title')
plt.ylabel('Ratings Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()

# Print the book with the maximum ratings count
max_ratings_book_title = df.loc[df['work_ratings_count'].idxmax(), 'title']
max_ratings_count = df['work_ratings_count'].max()
print(f"The book with the maximum number of ratings is '{max_ratings_book_title}' with {max_ratings_count} ratings.")
```

# Q7: Bucket the average_rating of books into 11 buckets [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0] with 0.5 decimal rounding (eg: average_rating 3.0 to 3.49 will fall in bucket 3.0). Plot bar graph to show total number of books in each rating bucket.

```python
# Define the bucket edges
bucket_edges = np.arange(0, 5.1, 0.5)

# Bucket the average_rating into the defined buckets
df['rating_bucket'] = pd.cut(df['average_rating'], bins=bucket_edges, right=False, labels=bucket_edges[:-1])

# Count the number of books in each rating bucket
bucket_counts = df['rating_bucket'].value_counts().sort_index()

# Create a bar plot to show the total number of books in each rating bucket
plt.figure(figsize=(10, 6))
bucket_counts.plot(kind='bar', color='lightblue')
plt.title('Number of Books in Each Rating Bucket')
plt.xlabel('Rating Bucket')
plt.ylabel('Number of Books')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print("Number of books in each rating bucket:")
for bucket, count in bucket_counts.items():
    print(f"Bucket {bucket} - {bucket + 0.5}: {count} books")
```
